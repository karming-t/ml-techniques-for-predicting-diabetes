```python
featured_unclean_df.rename(
    columns={
        "_AGEG5YR":"age_grp",
        "SEX":"sex",
        "_RACE":"race_cat",
        "_BMI5CAT":"bmi_grp",
        "_EDUCAG": "education",
        "_INCOMG":"income_grp",
        "_RFHLTH":"gen_health",
        "_PACAT1":"phy_activeness",
        "_SMOKER3":"smoke",
        "_RFDRHV5":"drink",
        "_RFHYPE5":"high_bp",
        "_RFCHOL":"high_chol",
        "CVDSTRK3":"stroke",
        "_MICHD":"heart_disease_attack",
        "CHCKIDNY":"kidney_disease",
        "DIABETE3":"diagnosed_diabetes",
    },
    inplace=True
)
featured_unclean_df.head(21)
```

```python
#total number of NaN
total_nan = featured_unclean_df.isna().sum()
print(total_nan)
print(featured_unclean_df.shape)
```

```python
# drop nan option
drop_nan_df = featured_unclean_df.dropna()
drop_nan_df.shape
```

```python
# the values in the rows shouldnt be containing zeroes
drop_nan_df = drop_nan_df[(drop_nan_df != 0).all(axis=1)]
#compare to the shape before this block to see how many rows contains value of zero
drop_nan_df.shape
```

```python
# remove rows with "dontknow/refused/missing" value labels for each feature
#AGE
drop_nan_df = drop_nan_df[drop_nan_df["age_grp"] != 14]
```

```python
#RACE
drop_nan_df = drop_nan_df[drop_nan_df["race_cat"] != 9]
```

```python
#EDUCATION
drop_nan_df = drop_nan_df[drop_nan_df["education"] != 9]
```

```python
#INCOME
drop_nan_df = drop_nan_df[drop_nan_df["income_grp"] != 9]
```

```python
#GENERAL HEALTH
drop_nan_df = drop_nan_df[drop_nan_df["gen_health"] != 9]
```

```python
#PHY ACTIVENESS
drop_nan_df = drop_nan_df[drop_nan_df["phy_activeness"] != 9]
```

```python
#SMOKING
drop_nan_df = drop_nan_df[drop_nan_df["smoke"] != 9]
```

```python
#DRINKING
drop_nan_df = drop_nan_df[drop_nan_df["drink"] != 9]
```

```python
#HIGH BP
drop_nan_df = drop_nan_df[drop_nan_df["high_bp"] != 9]
```

```python
#HIGH CHOL
drop_nan_df = drop_nan_df[drop_nan_df["high_chol"] != 9]
```

```python
#STROKE
drop_nan_df = drop_nan_df[~drop_nan_df["stroke"].isin([7,9])]
```

```python
#KIDNEY DISEASE
drop_nan_df = drop_nan_df[~drop_nan_df["kidney_disease"].isin([7,9])]
```

```python
#DIAGNOSED DIABETES - NAN VALUES
drop_nan_df = drop_nan_df[~drop_nan_df["diagnosed_diabetes"].isin([7,9])]
```

```python
#DIAGNOSED DIABETES - Gestational and Prediabetes Removal
drop_nan_df = drop_nan_df[~drop_nan_df["diagnosed_diabetes"].isin([2,4])]
```

```python
# DIAGNOSED DIABETES - Renumbering
drop_nan_df['diagnosed_diabetes'] = drop_nan_df['diagnosed_diabetes'].replace(3, 0)
```
